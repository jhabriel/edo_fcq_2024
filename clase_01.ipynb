```python
import numpy as np
import matplotlib.pyplot as plt
```

## Ejercicio 1.1.19, Zill 8va Ed.

```python
# Funci칩n
def X(t):
    return (np.exp(t) - 1) / (np.exp(t) - 2)

# Valores de variable independiente y dependiente
t_vals = np.linspace(-10, 10, 50)
X_vals = X(t_vals)

# Graficamos
fig, ax = plt.subplots()
ax.plot(t_vals, X_vals, "o-k")
ax.set(xlabel='t', ylabel='X(t)', title="Graficando inocentemente X(t)")
ax.grid()
plt.show()
```

Un simple an치lisis del denominador nos dice que la funci칩n $X(t) = \displaystyle{\frac{e^t - 1}{e^t - 2}}$ no est치 definida en $t = \text{ln} 2$.

```python
epsilon = 0.25
# valor peque침o alejado de la discontinuidad
t_vals_left = np.linspace(-10, np.log(2) - epsilon, 25)  # VI, valores a la izquierda de ln(2)
t_vals_right = np.linspace(np.log(2) + epsilon, 10, 25)  # VI, valores a la derecha de ln(2)

X_vals_left = X(t_vals_left)  # VD, valores a la izquierda de ln(2)
X_vals_right = X(t_vals_right)  # VD, valores a la derecha de ln(2)

# Graficamos
fig, ax = plt.subplots()
ax.plot(t_vals_left, X_vals_left, "o-b", label=r"$t \in (-\infty, \text{ln}~2)$")
ax.plot(t_vals_right, X_vals_right,"o-g", label=r"$t \in (\text{ln}~2, \infty)$")
ax.set(xlabel='t', ylabel='X(t)', title="Dominio de la funci칩n X(t) es " + r"$\mathbb{R} \setminus \{\text{ln}2\}$")
ax.legend()
ax.grid()
plt.show()
```

```python
## Gr치ficas de las soluciones de la EDO

fig, ax = plt.subplots(1, 2)

ax[0].plot(t_vals_left, X_vals_left, "-ob")
ax[0].set(xlabel='t', ylabel='X(t)', title="Soluci칩n X(t) en [-10, ln2-0.5]")

ax[1].plot(t_vals_right, X_vals_right,"-og")
ax[1].set(xlabel='t', ylabel='X(t)', title="Soluci칩n X(t) en [ln2+0.5, 10]")

plt.tight_layout()
plt.show()
```

## Ejercicio 21, Zill, 8Ed

```python
# Funci칩n
def P(t, c1):
    return (c1 * np.exp(t)) / (1 + c1 * np.exp(t))


# Analizamos el caso c1 >= 0

fig, ax = plt.subplots()
    
for c1 in [0, 1, 2, 3, 4]:

    # Valores de variable independiente y dependiente
    t_vals = np.linspace(-5, 5, 50)
    P_vals = P(t_vals, c1)
    
    # Graficamos
    ax.plot(t_vals, P_vals, label=f"c1 = {c1}")

ax.set(xlabel='t', ylabel='P(t)', title="P(t)")
ax.grid()
ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots()

# Como se puede ver, P(t) es discontinuo en ln(-1/c1) para todo c1 < 0.
for c1 in [-1, -2, -3, -4]:

    # Valores de variable independiente y dependiente
    t_vals = np.linspace(-5, 5, 50)
    P_vals = P(t_vals, c1)
    
    # Graficamos
    ax.plot(t_vals, P_vals, label=f"c1 = {c1}")

ax.set(xlabel='t', ylabel='P(t)', title="P(t)")
ax.grid()
ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots(1, 2)

# Los intervalos de validez de la EDO son los que se encuentran
# a la izquierda y derecha de la discontinuidad, respectivamente

for c1 in [-1, -2, -3, -4]:

    epsilon = 0.5

    t_vals_left = np.linspace(-5, np.log(-1/c1) - epsilon, 50)
    P_vals_left = P(t_vals_left, c1)
    ax[0].plot(t_vals_left, P_vals_left, label=f"c1 = {c1}")

    t_vals_right = np.linspace(np.log(-1/c1) + epsilon, 5, 50)
    P_vals_right = P(t_vals_right, c1)
    ax[1].plot(t_vals_right, P_vals_right, label=f"c1 = {c1}")


ax[0].set(xlabel='t', ylabel='P(t)', title="P(t)")
ax[0].grid()
ax[0].legend()

ax[1].set(xlabel='t', ylabel='P(t)', title="P(t)")
ax[1].grid()
ax[1].legend()

plt.tight_layout()
plt.show()
```
