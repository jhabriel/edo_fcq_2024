```python
import numpy as np
import matplotlib.pyplot as plt
```

## Ejericio 2.1.1, Zill, 9na Ed.

```python
def dx_dy(x, y):
    return x ** 2 - y ** 2

X, Y = np.meshgrid(np.linspace(-3, 3, 30), np.linspace(-3, 3, 30))
U = np.ones_like(X)
V = dx_dy(X, Y)

fig, ax = plt.subplots(figsize=(6, 6))

ax.quiver(X, Y, U, V, scale=120)

ax.plot([-2], [1], ".r", markersize=15)
ax.plot([3], [0], ".", markersize=15)
ax.plot([0], [2], ".", markersize=15)
ax.plot([0], [0], ".", markersize=15)


ax.set(xlabel="x", ylabel="y")
plt.show()
```

## Ejercicio 2.19

```python
# Campo direccional para la ecuacin autnoma
def dx_dy(y):
    return y - y ** 3

X, Y = np.meshgrid(np.linspace(0, 3, 30), np.linspace(1, 5, 30))
U = np.ones_like(X)
V = dx_dy(Y)

fig, ax = plt.subplots(figsize=(6, 6))

ax.quiver(X, Y, U, V, scale=400)

ax.set(xlabel="x", ylabel="y")
plt.show()
```

```python
Como se puede ver en la grfica, en el campo direccional de una EDO autnoma, las flechas coinciden a lo largo del eje x, pero varan en el eje y.
```

## Ejercicio 37, Zill, 9na Ed. 

\begin{equation}
\frac{dy}{dx} + 2y = f(x)
\end{equation}

\begin{equation}
    f(x) = 
        \begin{cases}
            1 & 0 \leq x \leq 3 \\
            0 & x > 3
        \end{cases}
\end{equation}

con $y(0) = 0$.

Al resolver utilizando factor integrante $P(x) = 2$, la condicin inicial $y(0)=0$ y la continuidad de la solucin en $x=3$, obtenemos:

\begin{equation}
y(x) = 
    \begin{cases}
        \frac{1}{2} (1 - e^{-2x}) & 0 \leq x \leq 3 \\
        \frac{1}{2} e^{2(3 - x)} - \frac{1}{2} e^{-2x} & x > 3
    \end{cases}
\end{equation}

```python
def y_left(x):
    return 0.5 * (1 - np.exp(-2 * x))

def y_right(x):
    return 0.5 * np.exp(2 * (3 - x)) - 0.5 * np.exp(-2 * x)


# 0 <= x <= 3
x_vals_left = np.linspace(0, 3, 50)
y_vals_left = y_left(x_vals_left)

# x > 3
x_vals_right = np.linspace(3.001, 10, 50)
y_vals_right = y_right(x_vals_right)

fig, ax = plt.subplots()
ax.plot(x_vals_left, y_vals_left, "o-b", linewidth=2, label=r"$x\in [0, 3]$")
ax.plot(x_vals_right, y_vals_right, "o-r", linewidth=2, label=r"$x\in (3, \infty)$")

ax.set(xlabel='x', ylabel='y(x)')
ax.grid()
ax.legend()
plt.show()
```
