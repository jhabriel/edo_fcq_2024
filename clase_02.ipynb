```python
import numpy as np
import matplotlib.pyplot as plt

from matplotlib import cm
from matplotlib.ticker import LinearLocator
```

## Ejercicio 1.2.1 (Zill Ed. 8)

```python
def y(x):
    return 1 / (1 - 4 * np.exp(-x))

eps = 0.25

x_vals_left = np.linspace(-4, -np.log(0.25) - eps, 25)
x_vals_right = np.linspace(-np.log(0.25) + eps, 6, 25)
y_vals_left = y(x_vals_left)
y_vals_right = y(x_vals_right)

fig, ax = plt.subplots()
ax.plot(x_vals_left, y_vals_left, "-b", linewidth=2)
ax.plot(x_vals_right, y_vals_right, "-b", linewidth=2)

ax.plot(0, -1/3, ".m", markersize=15)

ax.set(xlabel='x', ylabel='y(x)')
ax.grid()
plt.show()
```

## Ejercicio 1.2.2 (Zill Ed. 8)

```python
def y(x):
    return 1 / (1 - 0.5 * np.exp(-(x + 1)))

eps = 0.1

x_vals_left = np.linspace(-6, (-1 - np.log(2)) - eps, 25)
x_vals_right = np.linspace((-1 - np.log(2)) + eps, 3, 25)
y_vals_left = y(x_vals_left)
y_vals_right = y(x_vals_right)

fig, ax = plt.subplots()
ax.plot(x_vals_left, y_vals_left, "-b", linewidth=2)
ax.plot(x_vals_right, y_vals_right, "-b", linewidth=2)

ax.plot(-1, 2, ".m", markersize=15)

ax.set(xlabel='x', ylabel='y(x)')
ax.grid()
plt.show()
```

## Ejercicio 1.2.3 (Zill Ed. 8)

```python
def y(x):
    return 1 / (x ** 2 - 1)

eps = 0.1

x_vals_left = np.linspace(-4, -1 - eps, 25)
x_vals_middle = np.linspace(-1 + eps, 1 - eps, 10)
x_vals_right = np.linspace(1 + eps, 4, 25)
y_vals_left = y(x_vals_left)
y_vals_middle = y(x_vals_middle)
y_vals_right = y(x_vals_right)

fig, ax = plt.subplots()
ax.plot(x_vals_left, y_vals_left, "-b", linewidth=2)
ax.plot(x_vals_middle, y_vals_middle, "-b", linewidth=2)
ax.plot(x_vals_right, y_vals_right, "-b", linewidth=2)

ax.plot(2, 1/3, ".m", markersize=15)

ax.set(xlabel='x', ylabel='y(x)')
ax.grid()
plt.show()
```

En un PVI para una EDO de primer orden, para que una solucin particular $y(x)$ satisfaga la ODE, la grfica de $y(x)$ debe pasar por $(x_0, y_0)$. Es decir, el intervalo $I$ de la solucin **debe** contener a la condicin inicial.

## Ejercicio 1.2.7 (Zill Ed. 8)

```python
def x(t):
    return -np.cos(t) + 8 * np.sin(t)

t_vals = np.linspace(-np.pi, np.pi, 50)
x_vals = x(t_vals)

fig, ax = plt.subplots()
ax.plot(t_vals, x_vals, "-r", linewidth=2)

ax.plot(0, -1, ".m", markersize=15)  # x = -1 en t = 0
ax.plot(np.array([-1, 1]), np.array([-9, 7]), "--g", linewidth=2)  # x' = 8 en t = 0

ax.set(xlabel='t', ylabel='x(t)')
ax.grid()
plt.show()
```

En un PVI para una EDO de segundo orden, para que una solucin particular $y(x)$ satisfaga la ODE: (1) La grfica de $y(x)$ debe pasar por $(x_0, y_0)$, y (2) La pendiente en $(x_0, y_0)$ debe coincidir con el valor de la derivada asociada (a la otra) condicin inicial.

## Ejercicio 1.2.17 (Zill Ed. 8)

```python
def f(x, y):
    return np.cbrt(y) ** 2

def dfdy(x, y):
    return (2/3) * (1 / (np.cbrt(y)))

# Valores a la izquierda de y = 0 
a, b = -10, 10
c, d = -10, -0.1

X, Y = np.meshgrid(np.linspace(a, b, 100), np.linspace(c, d, 100))

f_vals = f(X, Y)
df_vals = dfdy(X, Y)

fig, ax = plt.subplots(1, 2, subplot_kw={"projection": "3d"}, figsize=[10, 10])
ax[0].plot_surface(X, Y, f_vals, cmap=cm.coolwarm, linewidth=0, antialiased=False)
ax[1].plot_surface(X, Y, df_vals, cmap=cm.coolwarm, linewidth=0, antialiased=False)

ax[0].set(xlabel="x", ylabel="y", title=r"$f = y'$")
ax[0].view_init(20, -10)

ax[1].set(xlabel="x", ylabel="y", title=r"$\partial f / \partial y$")
ax[1].view_init(20, -10)

plt.show()
```

```python
# Valores a la derecha de y = 0 
a, b = -10, 10
c, d = 0.1, 10

X, Y = np.meshgrid(np.linspace(a, b, 100), np.linspace(c, d, 100))

f_vals = f(X, Y)
df_vals = dfdy(X, Y)

fig, ax = plt.subplots(1, 2, subplot_kw={"projection": "3d"}, figsize=[10, 10])
ax[0].plot_surface(X, Y, f_vals, cmap=cm.coolwarm, linewidth=0, antialiased=False)
ax[1].plot_surface(X, Y, df_vals, cmap=cm.coolwarm, linewidth=0, antialiased=False)

ax[0].set(xlabel="x", ylabel="y", title=r"$f = y'$")
ax[0].view_init(20, -10)

ax[1].set(xlabel="x", ylabel="y", title=r"$\partial f / \partial y$")
ax[1].view_init(20, -10)

```

## Ejercicio 1.2.18 (Zill Ed. 8)

```python
def f(x, y):
    return (x * y) ** (1/2)

def df(x, y):
    return (1/2) * ((x * y) ** (1/2) / y)

# Para valores negativos de x e y
a, b = -10, -0.1
c, d = -10, -0.1

X, Y = np.meshgrid(np.linspace(a, b, 100), np.linspace(c, d, 100))

f_vals = f(X, Y)
df_vals = df(X, Y)

fig, ax = plt.subplots(1, 2, subplot_kw={"projection": "3d"}, figsize=[10, 10])
surf = ax[0].plot_surface(X, Y, f_vals, cmap=cm.coolwarm, antialiased=False)
surf = ax[1].plot_surface(Y, Y, df_vals, cmap=cm.coolwarm, antialiased=False)

ax[0].set(xlabel="x", ylabel="y", title=r"$f(x, y) = y'$")
ax[0].view_init(20, -10)

ax[1].set(xlabel="x", ylabel="y", title=r"$\partial f / \partial y$")
ax[1].view_init(20, -20)

plt.show()
```

```python
# Para valores positivos de x e y
a, b = 0.1, 10
c, d = 0.1, 10

X, Y = np.meshgrid(np.linspace(a, b, 100), np.linspace(c, d, 100))

f_vals = f(X, Y)
df_vals = df(X, Y)

fig, ax = plt.subplots(1, 2, subplot_kw={"projection": "3d"}, figsize=[10, 10])
surf = ax[0].plot_surface(X, Y, f_vals, cmap=cm.coolwarm, antialiased=False)
surf = ax[1].plot_surface(Y, Y, df_vals, cmap=cm.coolwarm, antialiased=False)

ax[0].set(xlabel="x", ylabel="y", title=r"$f(x, y) = y'$")
ax[0].view_init(20, 10)

ax[1].set(xlabel="x", ylabel="y", title=r"$\partial f / \partial y$")
ax[1].view_init(20, 10)

plt.show()
```

Las ecuaciones diferenciales de primer orden tienen una solucin nica si tanto $f(x, y) = y'$ como $\partial f / \partial y$ son continuos en esa regin.

## Ejercicio 1.12.6 (Zill 8va Ed)

```python
# Con coseno
def Tm_cos(t):
    A = 30  # 30 amplitud
    B = 24  # 24 periodo
    C = -12  # -12 desplazamiento horizontal
    D = 80  # 80 desplazamiento vertical
    return A * np.cos(2 * np.pi  / B * (t + C)) + D

# Con seno
def Tm_sin(t):
    A = 30  # 30 amplitud
    B = 24  # 24 periodo
    C = -6 # -6 desplazamiento horizontal
    D = 80  # 80 desplazamiento vertical
    return A * np.sin(2 * np.pi / B *  (t + C)) + D

t_vals = np.linspace(0, 48, 100)

fig, ax = plt.subplots()
ax.plot(t_vals, Tm_cos(t_vals), "-", linewidth=2)
ax.plot(t_vals, Tm_sin(t_vals), "o")

ax.set(xlabel='t [h]', ylabel='T_m [C]')
ax.grid()
plt.show()
```
